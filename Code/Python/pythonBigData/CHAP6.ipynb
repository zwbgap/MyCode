数据转换方法

```python

def Covd(x):
    if 'A' <= x <= 'A-':
        return '优'
    elif 'B' <= x <= 'B-':
        return '良'
    elif 'C' <= x <= 'C-':
        return '中'
    elif 'D' <= x <= 'D-':
        return '及格'
    elif x == 'E':
        return '不及格'
    else:
        return None
```

```python

df['C6'] = df['C5'].apply(Covd)
df['C7'] =df['C5'].map(Covd)
df.head(10)
#df
```

```python
df['性别'] = df['身份证号'].apply(lambda x: '女' if int(x[16]) % 2 == 0 else '男')
df.head(10)
```
